# JPMorgan Chase & Co. (JPM) Stock Historical Prices

```python
jpm_df = pd.read_csv("JPM.csv")
```

```python
jpm_df.head()
```

## 1. Line Plot - Stock Price Trend Over Time

```python
plt.figure(figsize=(10,5))
plt.plot(jpm_df.index, jpm_df['Close'], label='Closing Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price Trend Over Time')
plt.legend()
plt.grid()
plt.show()
```

## 2. Bar Chart - Comparing Closing Prices

```python
df_tail = jpm_df.tail(10)

plt.figure(figsize=(10,5))
plt.bar(df_tail.index.strftime('%Y-%m-%d'), df_tail['Close'], color='orange')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Closing Prices (Last 10 Days)')
plt.xticks(rotation=45)
plt.show()
```

## 3. Histogram - Distribution of Closing Prices

```python
plt.hist(jpm_df['Close'], bins=20, color='green', edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Closing Prices')
plt.show()
```

## 4. Scatter Plot - Relationship Between Volume and Closing Price

```python
plt.scatter(jpm_df['Open'], jpm_df['Close'], alpha=0.5, color='red')
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.title('Open vs Close Prices')
plt.show()
```

## 5. Box Plot - Stock Price Variations

```python
sns.boxplot(x=jpm_df['Close'])
plt.title('Boxplot of Closing Prices')
plt.show()
```

## 6. Pie Chart - Proportion of categories

```python
jpm_df["Price Category"] = pd.cut(
    jpm_df["Open"], 
    bins=5,  # Define 5 price range categories
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

# Plot pie chart for the price categories
plt.figure(figsize=(6, 6))
jpm_df["Price Category"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Pie Chart: Open Price Categories")
plt.ylabel("")  # Hide y-axis label
plt.show()
```
